## Pre-Processing

preprocessing data fitur menggunakan MinMaxScalar

$$X_{norm} = \frac{X_{i} - X_{min}}{X_{max} - X_{min}}$$

$X_i$ merupakan $i^{th}$ sample dari dataset. 

```python
scaler=MinMaxScaler()
```

```python
dfC=df.drop(columns=['label'])
dfL= pd.DataFrame(df, columns = ['label'])
```

```python
DataNorm=scaler.fit_transform(dfC.values)
```

```python
frameDataNorm=pd.DataFrame(DataNorm)
frameDataNorm.columns=['N', 'P', 'K', 'temperature', 'humidity', 'ph', 'rainfall']
```

```python
df_new = pd.concat([frameDataNorm, dfL], axis=1)
df_new
```

## Splitting Data

```python
# x = df.drop(['label'], axis = 1)
# y = df['label']
X = df_new.iloc[:,:-1].values 
y = df_new.iloc[:,-1].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, shuffle = True)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score,confusion_matrix,roc_auc_score
```

```python
df
```

## Membangun dan menerapkan Model Classifier

```python
from sklearn.naive_bayes import GaussianNB
#GNB
from sklearn.ensemble import RandomForestClassifier
gnb = GaussianNB()
gnb.fit(X_train,y_train)
gnb_pred = gnb.predict(X_test)
acc_gnb=round(accuracy_score(y_test, gnb_pred)*100, 2)
print(classification_report(y_test,gnb_pred))


acc_gnb
```

```python
model_accuracy = pd.DataFrame(columns=['Model','Accuracy'])
models = {
          "KNN" : KNeighborsClassifier(),
          "DT" : DecisionTreeClassifier(),
          'RFC' : RandomForestClassifier(n_estimators=50),
          "GNB" : GaussianNB(),

          }

for test, clf in models.items():
    clf.fit(X_train, y_train)
    y_pred = clf.predict(X_test)
    # acc = accuracy_score(y_test,y_pred)
    train_pred = clf.predict(X_train)
    train_acc = accuracy_score(y_train, train_pred)
    print("\n", test + ' scores')
    # print(acc)
    print(classification_report(y_test,y_pred))
    print('*' * 100,"\n")
    model_accuracy = model_accuracy.append({'Model': test, 'Train_acc': train_acc}, ignore_index=True)
```
