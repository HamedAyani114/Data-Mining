```python
from sklearn.preprocessing import LabelEncoder
# creating initial dataframe
rata_rata_overdue = ("0 - 30 days", "31 - 45 days", "46 - 60 days", "61 - 90 days", "> 90 days")
rata_rata_overdue_df = pd.DataFrame(rata_rata_overdue, columns=['rata_rata_overdue'])
# creating instance of labelencoder
labelencoder = LabelEncoder()
# Assigning numerical values and storing in another column
rata_rata_overdue_df['rata_rata_overdue_Cate'] = labelencoder.fit_transform(rata_rata_overdue_df['rata_rata_overdue'])
```
